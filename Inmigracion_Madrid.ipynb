```python
import pandas as pd
import numpy as np

# Visualización
# ----------------------------
import seaborn as sn
import matplotlib.pyplot as plt

#  Gestión de warnings
# ------------------------------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")

# Para ver todas las columnas
# ==================================
pd.options.display.max_columns = None
```

```python
df_nacionalidad = pd.read_excel('datos/inmigracion_nacionalid_madrid_2022.xlsx', skiprows=11)
df_nacionalidad.head()
```

```python
df_barrios = pd.read_excel('datos/inmigracion_por_barrios.xlsx', skiprows=9)
df_barrios.head()
```

```python
df_inmigracion = pd.read_csv('datos/Immigration_Madrid_2021.csv', index_col=0)
df_inmigracion.head()
```

> Hacemos EDA  y limpieza de los archihvos:
>
>- Tenemos datos para el 2021 y 2022 y decidimos quedarnos con los datos del 2021
> - Para el df de  nacionalidad, nos quedamos con el nº de habit. y el % de incremento.
>- En el df de distribución por barrios, nos quedamos con distritos en lugar de por barrios.

> ### Limpieza archivo Nacionalidad

```python
# nos quitamos las columnas que no queremos

df_nacionalidad.drop(['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 5','Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 10'], axis=1, inplace=True)
```

```python
# renombramos las columnas

columnas = {'Unnamed: 1':'Nacionalidad', 'Unnamed: 3':'nº_habitantes', 'Unnamed: 9':'Incremento(%)'}
df_nacionalidad.rename(columns=columnas, inplace=True)
```

```python
df_nacionalidad.head()
```

> ### Limpieza archivos Barrios

```python
# nos quitamos las columnas que no queremos

df_barrios.drop(['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 5','Unnamed: 6', 'Unnamed: 4', 'Unnamed: 7'], axis=1, inplace=True)
```

```python
# renombramos las columnas

columnas2 = {'Unnamed: 1':'Distritos', 'Unnamed: 8':'Nacionales', 'Unnamed: 9':'Otras_nacionalidades', 'Unnamed: 10': 'Extranjeros(%)'}
df_barrios.rename(columns=columnas2, inplace=True)
```

```python
# Eliminamos las filas dónde aparecen los barrios para quedarnos solo con los distritos

df_barrios['Distritos'].unique()
```

```python
# sacamos las filas por distritos 
df_distritos = df_barrios.loc[[0,7,15,22,29,36,43,50,59,67,75,83,91,98,105,115,122,128,132,137,146]]
```

```python
# ordenamos los índices
df_distritos.head().reset_index(drop=True)
```

> ### Limpieza archivo Inmigración

```python
# Cambiamos a español el idioma de las columnas

columnas3 = {'Gender':'Genero', 'Age':'Edad', 'Studies':'Estudios', 'Nationality': 'Nacionalidad', 'Administrative_situation':'Situación_administrativa', 'Working':'Trabaja' }
df_inmigracion.rename(columns=columnas3, inplace=True)
```

```python
df_inmigracion.head()
```

> ### Unimos los archivos

```python
# unimos los dos df que tinen la misma columna (Nacionalidad) pero OJO!! los datos no están sacados de la misma fuente por lo que no tienen un orde
df_final = df_inmigracion.merge(df_nacionalidad, how = 'outer')
```
